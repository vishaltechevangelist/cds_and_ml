### Replacing NaN
#### Mode for categorical value

```python
dt = dataset.copy()
dt
```

```python
cols = ['Gender', 'Area', 'Loan_Status']
```

```python
dt[cols] = dt[cols].fillna(dt[cols].mode().iloc[0])
dt
```

```python
dt[cols]
```

```python
dt.isnull().sum()
```

#### Mean for numerical value

```python
num_cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']
```

```python
dt[num_cols] = dt[num_cols].fillna(dt[num_cols].mean())
dt
```

```python
dt.isnull().sum()
```

### Label Encoding

```python
dt.dtypes
```

```python
dt[cols] = dt[cols].astype('category')
```

```python
dt.dtypes
```

```python
for columns in cols:
    dt[columns] = dt[columns].cat.codes
dt
```

### Train Test Split

```python
cleandata = dataset.dropna(subset=['Loan_Status'])
cleandata
```

```python
df = cleandata.copy()
df
```

```python
df[cols] = df[cols].fillna(df.mode().iloc[0])
df
```

```python
df.isnull().sum()
```

```python
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())
df
```

```python
df = df.drop(['Loan_ID'], axis=1)
df
```

```python
df = pd.get_dummies(df, drop_first=True)
df
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:, -1]
X, y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
X_train, X_test
```
